```python
# 2. Adding elements into the lists
# Creating a list with initial elements
my_list = [1, 2, 3, 'student', 'python']
# Creating a list using the list() constructor
my_list_2 = list([4, 5, 3.142, 'fun', 'cool'])
# Printing both lists
print(my_list)
print(my_list_2)
```

```python
# 3. Accessing elements of the list
my_list = [1, 2, 3, 'student', 'python']
print(my_list[:])       # Access all elements
print(my_list[3])       # Access the 4th element (index 3)
print(my_list[0:4])     # Access elements from index 0 to 3
print(my_list[::-1])    # Access elements in reverse order
print(my_list[3][2])    # Access a character from the string 'student' (index 2 → 'u')
```

```python
#Exmple: Changing Items
numbers = [4, 5, 6]
numbers[0] = 10
print(numbers)
```

```python
# Working with append() and extend()
my_list = [1, 2, 3, 'compiler', 'python']
my_list_2 = [4, 5, 3.142, 'fun', 'cool']
# append() adds the entire passed item as a single element
my_list.append([555, 12])  #append() → adds [555, 12] as one element
print(my_list)
print(len(my_list))
# extend() adds items one-by-one into the list
my_list_2.extend([555, 12]) #extend() → adds 555 and 12 separately
print(my_list_2)
print(len(my_list_2)) #len() shows the updated list size
```

```python
# Using insert(), concatenation, and multiplication operations on lists
# Insert an element at a specific index
my_list.insert(1, 'insert example')
print(my_list)
# Concatenate list with another list
print(my_list + ['just for example'])
# Multiply list (repeat elements)
print(my_list * 2)
# Print original list to show it remains unchanged after multiplication
print(my_list)
```

```python
# Working with extend(), del, remove(), pop(), and clear()
my_list = [1, 2, 3, 'compiler', 'python']
my_list_2 = [4, 5, 3.142, 'fun', 'cool']
# Extend the list by adding elements one-by-one
my_list_2.extend([555, 12, 'example'])
print(my_list_2)
# Delete element at index 5 using del
del my_list_2[5]
print(my_list_2)
# Remove a specific element by value (if present)
my_list_2.remove('example')
# Pop element at index 5 (also returns the removed value)
a = my_list_2.pop(5)
print('Popped Element:', a, ' | List remaining:', my_list_2)
# Clear the entire list
my_list_2.clear()
print(my_list_2)
```

```python
# Using list functions: len(), index(), count(), sorted(), sort(), copy()
my_list = [1, 2, 3, 'compiler', 'python']
my_list_2 = [4, 5, 3.142, 'fun', 'cool']
new = [5, 4, 12, 45, 3]
# Find the length of the list
print(len(my_list))
# Find the index of an element
print(my_list.index('compiler'))
# Count how many times an element appears in the list
print(my_list.count('compiler'))
# sorted() returns a new sorted list, original remains unchanged
print(sorted(new))
print(new)
# sort() modifies the list in place (reverse order)
new.sort(reverse=True)
print(new)
# Copy list elements into a new list
new_list = my_list.copy()
print(new_list)
```

*   Can contain different data types (ints, strings, floats, etc.)

```python
#Example-1:
my_tuple = (1, 2, 3, "python", 4.5)
print(my_tuple)
```

```python
#Example-tuple
days = ("Mon", "Tue", "Wed")
print(type(days))   # <class 'tuple'>
```

```python
#Tuple Slicing
numbers = (10, 20, 30, 40, 50)
print(numbers[1:4])   # (20, 30, 40)
```

```python
# 1. Initializing tuples
# Tuple with multiple elements
my_tuple = (1, 2, 3)
# Another tuple using the tuple() constructor
my_tuple_2 = tuple(('python', 'for', 'compiler'))
# Printing the tuples
print(my_tuple)     # Output: (1, 2, 3)
print(my_tuple_2)   # Output: ('python', 'for', 'compiler')
# Tuple with a single element
my_tuple_3 = 'example',  # Note the comma at the end; without it, this is just a string
# Checking the type
print(type(my_tuple_3))  # Output: <class 'tuple'>
```

```python
# 2. Accessing elements of a tuple
# Tuple with numbers
my_tuple = (1, 2, 3)
# Tuple with strings
my_tuple_2 = ('python', 'for', 'compiler')
# Tuple containing a list
my_tuple_3 = (1, 2, 3, ['hindi', 'python'])
# Accessing the first element of my_tuple
print(my_tuple[0])   # Output: 1   #[] to access individual elements.
# Accessing all elements (slicing)
print(my_tuple[:])   # Output: (1, 2, 3) [:] to get a copy of the tuple.
# Accessing the second element inside the list within the tuple
print(my_tuple_3[3][1])  # Output: 'python'
```

**`5.Control Flow`**
*   Parts of the code where the sequence must alter — flow control comes to the rescue.
*  Control flow in Python refers to the order in which statements in a program are executed.
*  It allows your program to make decisions, repeat actions, and choose different paths based on conditions.

`*   `Why is control flow important? `
*   Because without it, a program would run line-by-line from top to bottom with no decisions or loops.

**`Main Control Flow`**
| Category              | Statements                  | Purpose                                            |
| --------------------- | --------------------------- | -------------------------------------------------- |
| **Decision Making**   | `if`, `elif`, `else`        | Run code only when certain conditions are true     |
| **Loops**             | `for`, `while`              | Repeat blocks of code multiple times               |
| **Jump/Loop Control** | `break`, `continue`, `pass` | Stop loops, skip iterations, or leave placeholders |


1. **`Decision Making`**

**`a. if Statement`**
*  Used when you want to execute a block of code only if a condition is true.

```python
#Syntax:
    #if condition:
        # code to execute when condition is true
```

```python
#Example-if statement
age = 23
if age >= 18:
    print("You are eligable.")
```

b. elif (Else If)
*   Used when you want to check multiple conditions.

```python
#Syntax
#if condition1:
    # executes if condition1 is true
#elif condition2:
    # executes if condition2 is true
```

```python
#Example: elif statement
score = 73
if score >= 90:
    print("Grade: A")
elif score >= 75:
    print("Grade: B")
elif score >= 50:
    print("Grade: C")
#Checks conditions in order until one is true.
```

c. else Statement
*   Used when all previous conditions are false.

```python
#Syntax:
#if condition:
    # code
#else:
    # code if condition is false
```

```python
#Example- else statement
temperature = 10
if temperature > 25:
    print("Hot day")
else:
    print("Cold day")
#Runs the else block because the condition is false.
```

```python
#Example: if..elif..else
a = 10
b = 15

if a == b:
    print('They are equal')
elif a > b:
    print('a is larger')
else:
    print('b is larger')
```

```python
#Example: Nested if-else statements
n=int(input("Enter a number:"))
if n>=0:
    if n==0:
        print("Entered number  is zero:")
    else:
        print("Entered number is positive:")
else:
        print('Entered number is negative')
```

**`2. Loops`**
*   In Python, loops are used to execute a block of code repeatedly as long as a certain condition is true or for each item in a sequence.
*   Loops are used when we want to execute a certain set of statements for different inputs several times
*   Loops help automate repetitive tasks, which saves time and avoids writing the same code multiple times.

**`Python mainly has two types of loops:`**
1.   for loop
*   Used to iterate over a sequence (like a list, tuple, string, or range).
*   Executes the block of code for each item in the sequence.

```python
#Syntax
#for variable in sequence:
    # code to execute
```

```python
#Example
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

**`2. while loop`**
*  Repeats a block of code as long as a condition is True.

```python
#Syntax
#while condition:
    # code to execute
```

```python
#Example
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
#while loop
i=1
sum=0
while i<=10:
    sum+=i
    i+=1
    print(sum)
```

**`Note:`**
*   Python does not have a built-in do...while loop like some other languages (C, C++, Java).
*   However, you can simulate a do...while loop using a while True: loop with a break at the end based on a condition.

```python
#Simulating do...while in Python
#Syntax:
#while True:
    # code to execute at least once
    #...
    #if not condition:
        #break
```

```python
# Example: Ask user to enter a positive number at least once
while True:
    num = int(input("Enter a positive number: "))
    if num > 0:
        print("Thank you!")
        break
    else:
        print("Try again. Number must be positive.")
```

**`3. Jump Statements / Loop Control Statements`**
*   Python provides three main statements to control loops:

b. break
*  Stops the loop immediately, regardless of the loop condition.
*   Useful when you want to exit early.

```python
#Example
for i in range(1, 10):
    if i == 5:
        break  # stop the loop
    print(i)
```

```python
#Example
count = 0
while True:
    print(count)
    count += 1
    if count == 3:
        break  # exit the loop when count is 3
```

**`b) continue`**
*   Skips the current iteration and moves to the next iteration.
*   Useful when you want to skip certain cases but continue looping.

```python
#Example
for i in range(1, 6):
    if i == 3:
        continue  # skip 3
    print(i)
```
